```python
import cv2
```

```python
from google.colab import files
img = files.upload()
```

```python
img=cv2.imread('MicrosoftTeams-image.png')
img
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
imgGray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(img.shape)
print(imgGray.shape)

cv2_imshow(imgGray)
```

```python
imgblur=cv2.GaussianBlur(imgGray, (7,7),0)
cv2_imshow(imgblur)
```

```python
imgCanny=cv2.Canny(imgblur, 100, 200)
cv2_imshow(imgCanny)
```

```python
wid=int(input("Enter width: "))
hei=int(input("Enter height: "))

imgresize=cv2.resize(img, (wid,hei))

print(imgresize.shape)
cv2_imshow(imgresize)
```

```python
imgcrop=img[280:500, 250: 400]
cv2_imshow(imgcrop)
```
